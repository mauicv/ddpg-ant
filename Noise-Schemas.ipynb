```python
from src.noise import OUNoise, LinearSegmentNoiseND, NormalNoise, SmoothNoiseND

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import time
```

```python
SIGMA = 0.15
THETA = 0.2
STEPS = 1000
NUM_INT_POINTS = int(STEPS/10)
```

```python
def plot_orb(noise_process):
    orb = []

    for i in range(STEPS):
        orb.append(noise_process())

    ax = plt.axes(projection='3d')
    orb = np.array(orb)
    orb.shape
    ax.plot3D(orb[:,0], orb[:,1], orb[:,2], 'red')
```

```python
normal_noise = NormalNoise(
    dim=3,
    sigma=SIGMA)

plot_orb(normal_noise)
```

```python
ou_noise = OUNoise(
    dim=3,
    sigma=SIGMA,
    theta=THETA,
    dt=1e-2)

plot_orb(ou_noise)
```

```python
linear_segment_noise = LinearSegmentNoiseND(
    steps=STEPS,
    dim=3,
    sigma=SIGMA,
    num_interp_points=NUM_INT_POINTS)

plot_orb(linear_segment_noise)
```

```python
smooth_noise = SmoothNoiseND(
    steps=STEPS,
    dim=3,
    sigma=SIGMA,
    num_interp_points=NUM_INT_POINTS)

plot_orb(smooth_noise)
```
